Part.2.D.4-recursion 中「变量的作用域」里面一句很重要的提示:

> 定义如此,但通畅程序员们会严格遵守一条原则:
>> 在函数内部绝对不调用全局变量。即使是必须改变全局变量，也只能通过函数的返回值在函数外改变全局变量。

这个介绍是在其他地方讲解这个知识点的时候，不会特地说明的一点。但是，从自己入门的经历来看，这个习惯的介绍，才是避免很多逻辑混乱、踩进很多坑的关键。如果这些基础要自己去慢慢摸索，那么入门的门槛会高出很多，让很多人产生怀疑自己的困扰，很可能有一部分人是这样放弃的。

其实，学习的结果，并不是为了掌握「核心的基础知识」，而是「形成自己的模块化工具」，这些原则、经验，其实就是模块化经验，这个才是友好的学习材料。

正如「排列组合」一样，基础的知识点，可以形成很多组合，好比初中数学里面的几何证明，即欧式几何，有的公理和定理其实不多，但是可以形成「无数的」的题目，但是总结一下会发现，几乎所有的题目，都有一些固定的模式，而这些模式是在基础知识点之上建立的。倘若直接从基础定理去证明，是很漫长的思考路径，但是如果使用「组合好后的、具有经验性质、模块化的模型」，那么思考路径就会有针对性、速度性。

再看一遍，又能找到新的体会，发现教程项目里面的亮点（自己第一遍是没有能力吸收这些内容的）。

其实，项目里面强调的「读书百遍其义自见」，就是这个意思的总结。第一遍是为了搭建框架，以后的每一遍过目，都是丰富、延展出更多的知识网络，从而达到知识的真正世界里面。
